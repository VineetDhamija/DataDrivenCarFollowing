```python
import numpy as np 
import pandas as pd 
import warnings
warnings.filterwarnings("ignore")
```

```python
from pathlib import Path
p = Path().cwd()
print(f"original File path: {p}")
stringpath= str(p)[0:str(p).rfind('\\')] + '\\data'
print(f"Updated File Path to pick from Data Library: { stringpath}")
ngsimfile=stringpath + '/' + 'Next_Generation_Simulation__NGSIM__Vehicle_Trajectories_and_Supporting_Data.csv'
ngsim = pd.read_csv(ngsimfile,low_memory=False)
print(f"{ngsim.duplicated().sum()} duplicate values have been removed")
ngsim.drop_duplicates(inplace=True)
```

```python
#filtered_ngsim= ngsim[((ngsim['Location'] == 'us-101') | (ngsim['Location'] == 'i-80')) & (ngsim['Preceding']>0) & (ngsim['Following']>0) ]
filtered_ngsim= ngsim[((ngsim['Location'] == 'us-101') | (ngsim['Location'] == 'i-80')) ]
filtered_ngsim= filtered_ngsim.drop(columns=['Movement', 'Direction','Section_ID','Int_ID','D_Zone','O_Zone','Lane_ID','Following','v_Width','Total_Frames','Frame_ID'])
```

```python
'''
change the below to metrics, 
field, current 

v_length=feet
Space_Headway=feet

v_vel=feet/second
v_acc=feet/second square


Time_Headway = seconds
Relative_Time= Global Time - Min(Global Time) +1

'''
#converting the Feet to Metric System
filtered_ngsim['v_length']= filtered_ngsim['v_length']/0.3048
filtered_ngsim['Space_Headway']= filtered_ngsim['Space_Headway']/0.3048
filtered_ngsim['v_Vel']= filtered_ngsim['v_Vel']/0.3048
filtered_ngsim['v_Acc']= filtered_ngsim['v_Acc']/0.3048
filtered_ngsim['Relative_Time']= filtered_ngsim['Global_Time'] - filtered_ngsim['Global_Time'].min() + 1
```

```python
filtered_ngsim[((filtered_ngsim['Space_Headway'] == 0 ) & (filtered_ngsim['Preceding'] > 0 )) ]
```

```python
filtered_ngsim['Preceding_Vehicle_Class'] = np.NaN
filtered_vClass= filtered_ngsim[['Vehicle_ID','v_Class','Location']].drop_duplicates().sort_values('v_Class')
```

```python
filtered_vClass
```

```python
v_Class_M_I     =set(filtered_vClass[(filtered_vClass['v_Class'] == 1 ) & (filtered_vClass['Location'] == 'i-80')   ]['Vehicle_ID'])
v_Class_C_I     =set(filtered_vClass[(filtered_vClass['v_Class'] == 2 ) & (filtered_vClass['Location'] == 'i-80')   ]['Vehicle_ID'])
v_Class_HV_I    =set(filtered_vClass[(filtered_vClass['v_Class'] == 3 ) & (filtered_vClass['Location'] == 'i-80')   ]['Vehicle_ID'])
v_Class_M_U     =set(filtered_vClass[(filtered_vClass['v_Class'] == 1 ) & (filtered_vClass['Location'] == 'us-101') ]['Vehicle_ID'])
v_Class_C_U     =set(filtered_vClass[(filtered_vClass['v_Class'] == 2 ) & (filtered_vClass['Location'] == 'us-101') ]['Vehicle_ID'])
v_Class_HV_U    =set(filtered_vClass[(filtered_vClass['v_Class'] == 3 ) & (filtered_vClass['Location'] == 'us-101') ]['Vehicle_ID'])
```

```python
filtered_ngsim["v_Class_Name"] = filtered_ngsim["v_Class"].map({1:"Motorcycle", 2: "Car", 3: "Heavy Vehicle"})
```

```python
for i in filtered_ngsim.index:
    if filtered_ngsim['Location'][i] == 'us-101':
        if filtered_ngsim['Preceding'][i] in v_Class_C_U:
            filtered_ngsim['Preceding_Vehicle_Class'][i] = 'Car'
        elif filtered_ngsim['Preceding'][i] in v_Class_HV_U:
            filtered_ngsim['Preceding_Vehicle_Class'][i] = 'Heavy Vehicle'
        elif filtered_ngsim['Preceding'][i] in v_Class_M_U:
            filtered_ngsim['Preceding_Vehicle_Class'][i] = 'Motorcycle'
    elif filtered_ngsim['Location'][i] == 'i-80':
        if filtered_ngsim['Preceding'][i] in v_Class_C_I:
            filtered_ngsim['Preceding_Vehicle_Class'][i] = 'Car'
        elif filtered_ngsim['Preceding'][i] in v_Class_HV_I:
            filtered_ngsim['Preceding_Vehicle_Class'][i] = 'Heavy Vehicle'
        elif filtered_ngsim['Preceding'][i] in v_Class_M_I:
            filtered_ngsim['Preceding_Vehicle_Class'][i] = 'Motorcycle'
```

```python
filtered_ngsim['Vehicle_combination']= filtered_ngsim['Preceding_Vehicle_Class'] + '-' + filtered_ngsim['v_Class_Name']
```

```python
stringpath= str(p)[0:str(p).rfind('\\')] + '\\data'
ngsimfilteredfile=stringpath + '\\' + 'Cleaned_NGSIM_Data_Part_1.csv'
filtered_ngsim.to_csv(ngsimfilteredfile, index = False)
```
